# Linear and Qubic Graph

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(10) + 1
y = - x

plt.plot(x / 10, (y / 10) + 1 )
x
```

```python
def cubic_func(x):
  return - x**3 + x ** 2 + x
x2 = np.arange(1, 101) - 50
y2 = [(cubic_func(a) + 100_000) / 200_000 for a in x2]
plt.plot((x2 + 50) / 100, y2)
```
